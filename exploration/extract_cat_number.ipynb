```python
PATH_AOI = "../../../Downloads/AOI.shp"
gdf_AOI = gpd.read_file(PATH_AOI)
result = []
import re
for index in ID:
    PATH = "../../../Downloads/cc{}_shp/CC{}_vl_parcel_poly.shp".format(index, index)
    
    gdf = gpd.read_file(PATH)
    ag = gdf[gdf["PT_TYPE"] == "P"]
    
    pbar =tqdm(total=len(ag))
    
    for pt_code in ag.PT_CODE.unique():
        sample = ag[ag["PT_CODE"] == pt_code]
        #pbar.update(1)
        id = sample.index[0]
        if gdf_AOI["geometry"][0].contains(sample["geometry"][id]) and sample["geometry"][id].area > 850:
            print(sample["geometry"][id].area)
            result.append(sample["geometry"][id].area)
            path = "../test_dump/{}_{}_{}.shp".format(sample.loc[id, "PT_TYPE"], re.sub("/", "-", sample.loc[id, "PT_CODE"]), sample.loc[id, "PT_CCAT"])
            sample.to_file(path , driver='ESRI Shapefile')

result
```

```python
import pandas as pd
import plotly.express as px
data = pd.DataFrame({
    'Values': result
})
data = data[data['Values'] < 4500]
# Create histogram using Plotly Express
fig = px.histogram(data, x='Values', title='Histogram Example')

# Show the plot
fig.show()
```

```python
import os

base_PATH = "../test_dump"
files = [file for file in os.listdir(base_PATH) if file.endswith(".shp")]

result = gpd.GeoDataFrame(columns=['name','geometry'])
pbar = tqdm(total=len(files))
for file in files:
    gdf_single = gpd.read_file("{}/{}".format(base_PATH, file))
    for index, _ in gdf_single.iterrows():
        result.loc[len(result)] = [re.sub(".shp", "", file), gdf_single["geometry"][index]]
    pbar.update(1)

result.set_geometry(col='geometry', inplace=True)

result.to_file("../complete.geojson", driver="GeoJSON")

print(len(result))




```
